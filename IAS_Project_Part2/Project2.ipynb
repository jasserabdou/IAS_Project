## **3. Data Preparation & Preprocessing**
##### Initial Data Preparation
- Create deep copy of original dataset for preprocessing
- Handle missing values (forward/backward fill outdoor temperature)
- Fix problematic values in satisfaction column (convert to numeric)

##### DateTime Handling
- Convert date and time columns to proper datetime format
- Handle invalid datetime values using sequential timestamps
- Sort data chronologically by datetime

##### Outlier Treatment
- Cap outliers using percentile-based bounds (1st and 99th percentiles)
- Apply IQR method (Q1-1.5*IQR, Q3+1.5*IQR) to define boundaries
- Preserve data distribution by capping instead of removing outliers

```python
import os
import numpy as np
from scipy.signal import savgol_filter
```

```python
# Create a deep copy for preprocessing
df_processed = df.copy(deep=True)
```

```python
# Handle missing values
print("Handling missing values...")
# Forward fill missing outdoor temperature values
df_processed["Outside temp"] = df_processed["Outside temp"].ffill().bfill()
```

```python
# Handle problematic Satisfaction column
print("Handling Satisfaction column...")
# Check if column exists
if "Satisfaction" in df_processed.columns:
    # Try to convert directly
    try:
        df_processed["Satisfaction"] = pd.to_numeric(
            df_processed["Satisfaction"], errors="raise"
        )
    except ValueError as e:
        print(f"Found non-numeric values in Satisfaction column: {e}")
        # Check which rows have issues
        mask = (
            pd.to_numeric(df_processed["Satisfaction"], errors="coerce").isna()
            & df_processed["Satisfaction"].notna()
        )
        problematic_values = df_processed.loc[mask, "Satisfaction"].unique()
        print(f"Problematic values: {problematic_values}")
        # Replace problematic values with NaN then convert
        df_processed["Satisfaction"] = pd.to_numeric(
            df_processed["Satisfaction"], errors="coerce"
        )
        # Fill NaN values using forward fill and backward fill
        df_processed["Satisfaction"] = df_processed["Satisfaction"].ffill().bfill()
else:
    print("Satisfaction column not found in dataset")
```

```python
# Convert DateTime and sort chronologically
print("Converting datetime and sorting...")
# Handle formatting issues in Date and Time columns
df_processed["Date"] = (
    df_processed["Date"].astype(str).str.replace(r"(\d+/\d+/\d+).*", r"\1", regex=True)
)
df_processed["Time"] = df_processed["Time"].astype(str)
```

```python
# Try to create DateTime
try:
    df_processed["DateTime"] = pd.to_datetime(
        df_processed["Date"] + " " + df_processed["Time"],
        format="%d/%m/%Y %H:%M",
        errors="coerce",
    )

    # Check for any NaT values
    nat_count = df_processed["DateTime"].isna().sum()
    if nat_count > 0:
        print(
            f"Warning: {nat_count} rows have invalid datetime values. Using sequential datetimes."
        )
        # For rows with invalid datetime, create sequential values
        valid_rows = df_processed["DateTime"].notna()
        if valid_rows.any():
            # Get first valid datetime to use as reference
            first_valid_dt = df_processed.loc[valid_rows, "DateTime"].min()
        else:
            # If no valid datetimes, use current time
            first_valid_dt = pd.Timestamp.now().normalize()

        # Create sequential datetimes for problematic rows
        invalid_idx = df_processed.index[~valid_rows]
        for i, idx in enumerate(invalid_idx):
            df_processed.loc[idx, "DateTime"] = first_valid_dt + pd.Timedelta(
                minutes=15 * i
            )
except Exception as e:
    print(f"Error in datetime conversion: {e}")
    # Create sequential datetimes as fallback
    start_time = pd.Timestamp.now().normalize()
    df_processed["DateTime"] = pd.date_range(
        start=start_time, periods=len(df_processed), freq="15min"
    )

# Sort by DateTime
df_processed = df_processed.sort_values("DateTime")
```

```python
#  Handle outliers using capping at percentiles
print("Handling outliers...")
target_cols = ["Indoor_temperature_room", "Humidity", "Outside temp", "CO2_room"]
for col in target_cols:
    # Skip if column doesn't exist
    if col not in df_processed.columns:
        print(f"Column {col} not found, skipping outlier handling for it")
        continue

    # Convert to numeric if needed
    if df_processed[col].dtype == object:
        df_processed[col] = pd.to_numeric(df_processed[col], errors="coerce")

    # Calculate percentiles and IQR
    Q1 = df_processed[col].quantile(0.01)  # 1st percentile
    Q3 = df_processed[col].quantile(0.99)  # 99th percentile
    IQR = Q3 - Q1

    # Cap outliers instead of removing them
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_processed[col] = np.where(
        df_processed[col] < lower_bound, lower_bound, df_processed[col]
    )
    df_processed[col] = np.where(
        df_processed[col] > upper_bound, upper_bound, df_processed[col]
    )
```

## **4. Feature Engineering**
##### Time Features:
- Extract hour, day, month, day_of_week from datetime
- Create binary daytime indicator (7AM-7PM)
- Generate cyclical time features (sin/cos transformations)

##### Lag Features:
- Create temperature and humidity lags at multiple intervals
- Short term: 1, 3, 6 intervals (15min each)
- Long term: 12, 24 intervals (3h, 6h)

##### Statistical Features:
- Calculate rolling means (windows: 4, 12, 24, 48)
- Compute rolling standard deviations for volatility
- Generate rate of change features (1h difference)
- Create exponentially weighted moving averages

##### Physical Relationship Features:
- Calculate indoor-outdoor temperature difference
- Generate temperature-humidity interactions
- Compute vapor pressure for indoor and outdoor conditions
- Calculate vapor pressure deficit and heat index

##### Signal Smoothing
- Apply Savgol filter (window_length=7, polyorder=2)
- Create smoothed versions of key measurements
- Implement fallback to moving averages when needed

##### Missing Value Handling
- Forward fill NaNs to maintain temporal consistency
- Backward fill any remaining leading NaNs
- Replace infinities with large finite values
- Verify complete dataset with final NaN check

##### Save Processed Dataset
- Export cleaned and feature-rich dataset to CSV

```python
# Extract basic time features
print("Extracting time features...")
df_processed["hour"] = df_processed["DateTime"].dt.hour
df_processed["day"] = df_processed["DateTime"].dt.day
df_processed["month"] = df_processed["DateTime"].dt.month
df_processed["day_of_week"] = df_processed["DateTime"].dt.dayofweek
df_processed["is_daytime"] = (
    (df_processed["hour"] >= 7) & (df_processed["hour"] <= 19)
).astype(int)

# Create cyclical time features
print("Creating cyclical time features...")
df_processed["hour_sin"] = np.sin(df_processed["hour"] * (2 * np.pi / 24))
df_processed["hour_cos"] = np.cos(df_processed["hour"] * (2 * np.pi / 24))
df_processed["day_of_week_sin"] = np.sin(df_processed["day_of_week"] * (2 * np.pi / 7))
df_processed["day_of_week_cos"] = np.cos(df_processed["day_of_week"] * (2 * np.pi / 7))
df_processed["month_sin"] = np.sin(df_processed["month"] * (2 * np.pi / 12))
df_processed["month_cos"] = np.cos(df_processed["month"] * (2 * np.pi / 12))
```

```python
# Create lag features
print("Creating lag features...")
for lag in [1, 3, 6, 12, 24]:  # 15-minute intervals, so lag 24 = 6 hours
    df_processed[f"temp_lag_{lag}"] = df_processed["Indoor_temperature_room"].shift(lag)
    df_processed[f"humidity_lag_{lag}"] = df_processed["Humidity"].shift(lag)

# Create rolling statistics
print("Creating rolling window features...")
for window in [4, 12, 24, 48]:  # Different time windows
    # Mean features
    df_processed[f"temp_rolling_mean_{window}"] = (
        df_processed["Indoor_temperature_room"]
        .rolling(window=window, min_periods=1)
        .mean()
    )
    df_processed[f"humidity_rolling_mean_{window}"] = (
        df_processed["Humidity"].rolling(window=window, min_periods=1).mean()
    )

    # Volatility features
    df_processed[f"temp_rolling_std_{window}"] = (
        df_processed["Indoor_temperature_room"]
        .rolling(window=window, min_periods=1)
        .std()
    )
    df_processed[f"humidity_rolling_std_{window}"] = (
        df_processed["Humidity"].rolling(window=window, min_periods=1).std()
    )
```

```python
# Create rate of change features
print("Creating rate of change features...")
df_processed["temp_rate_of_change_1h"] = df_processed["Indoor_temperature_room"].diff(
    4
)  # 4 records = 1 hour
df_processed["humidity_rate_of_change_1h"] = df_processed["Humidity"].diff(4)

# Create exponentially weighted features
print("Creating exponentially weighted features...")
df_processed["temp_ewm_1h"] = (
    df_processed["Indoor_temperature_room"].ewm(span=4, min_periods=1).mean()
)
df_processed["humidity_ewm_1h"] = (
    df_processed["Humidity"].ewm(span=4, min_periods=1).mean()
)
```

```python
# Create physical relationship features
print("Creating environmental physics features...")
# Temperature difference (indoor vs outdoor) - crucial for thermal transfer
df_processed["temp_diff_in_out"] = (
    df_processed["Indoor_temperature_room"] - df_processed["Outside temp"]
)

# Safety check for Outdoor_relative_humidity_Sensor column
if "Outdoor_relative_humidity_Sensor" not in df_processed.columns:
    print("Warning: Outdoor_relative_humidity_Sensor not found, creating placeholder")
    df_processed["Outdoor_relative_humidity_Sensor"] = (
        50.0  # Default value as placeholder
    )

# Interaction features
df_processed["temp_humid_interaction"] = (
    df_processed["Indoor_temperature_room"] * df_processed["Humidity"] / 100
)
df_processed["out_temp_humid_interaction"] = (
    df_processed["Outside temp"]
    * df_processed["Outdoor_relative_humidity_Sensor"]
    / 100
)

# Vapor pressure calculation (better humidity representation)
# Handle potential division by zero or negative values
df_processed["indoor_vapor_pressure"] = (
    (df_processed["Humidity"] / 100)
    * 6.112
    * np.exp(
        (17.67 * df_processed["Indoor_temperature_room"])
        / (df_processed["Indoor_temperature_room"].clip(lower=0.1) + 243.5)
    )
)
df_processed["outdoor_vapor_pressure"] = (
    (df_processed["Outdoor_relative_humidity_Sensor"] / 100)
    * 6.112
    * np.exp(
        (17.67 * df_processed["Outside temp"])
        / (df_processed["Outside temp"].clip(lower=0.1) + 243.5)
    )
)

# Vapor pressure deficit (indicates moisture transfer potential)
df_processed["vapor_pressure_deficit"] = (
    df_processed["outdoor_vapor_pressure"] - df_processed["indoor_vapor_pressure"]
)

# Heat index (simplified feels-like temperature)
df_processed["heat_index"] = (
    -42.379
    + 2.04901523 * df_processed["Indoor_temperature_room"]
    + 10.14333127 * df_processed["Humidity"]
    - 0.22475541 * df_processed["Indoor_temperature_room"] * df_processed["Humidity"]
)
```

```python
# Apply Savgol filter for noise reduction
print("Applying Savgol filter for noise reduction...")
# Ensure enough data points for the filter
if len(df_processed) >= 7:  # Need at least window_length points
    try:
        # Use a window length of 7 for better smoothing
        df_processed["Indoor_temperature_room_smoothed"] = savgol_filter(
            df_processed["Indoor_temperature_room"], window_length=7, polyorder=2
        )
        df_processed["Humidity_smoothed"] = savgol_filter(
            df_processed["Humidity"], window_length=7, polyorder=2
        )
        df_processed["CO2_room_smoothed"] = savgol_filter(
            df_processed["CO2_room"], window_length=7, polyorder=2
        )
        df_processed["Outside_temp_smoothed"] = savgol_filter(
            df_processed["Outside temp"], window_length=7, polyorder=2
        )
    except Exception as e:
        print(f"Error applying Savgol filter: {e}")
        # Fallback to simple moving average
        df_processed["Indoor_temperature_room_smoothed"] = (
            df_processed["Indoor_temperature_room"]
            .rolling(window=7, min_periods=1)
            .mean()
        )
        df_processed["Humidity_smoothed"] = (
            df_processed["Humidity"].rolling(window=7, min_periods=1).mean()
        )
        df_processed["CO2_room_smoothed"] = (
            df_processed["CO2_room"].rolling(window=7, min_periods=1).mean()
        )
        df_processed["Outside_temp_smoothed"] = (
            df_processed["Outside temp"].rolling(window=7, min_periods=1).mean()
        )
else:
    print("Not enough data points for Savgol filter, using moving average instead")
    # Fallback to simple moving average
    df_processed["Indoor_temperature_room_smoothed"] = (
        df_processed["Indoor_temperature_room"].rolling(window=3, min_periods=1).mean()
    )
    df_processed["Humidity_smoothed"] = (
        df_processed["Humidity"].rolling(window=3, min_periods=1).mean()
    )
    df_processed["CO2_room_smoothed"] = (
        df_processed["CO2_room"].rolling(window=3, min_periods=1).mean()
    )
    df_processed["Outside_temp_smoothed"] = (
        df_processed["Outside temp"].rolling(window=3, min_periods=1).mean()
    )
```

```python
# Handle NaN values created during feature engineering
print("Handling missing values created during feature engineering...")
# Count NaNs before handling
nan_count_before = df_processed.isna().sum().sum()
print(f"NaN values before handling: {nan_count_before}")

# First, forward fill remaining NaNs (keeps temporal consistency)
df_processed = df_processed.ffill()
# Then, backward fill any remaining NaNs at the start
df_processed = df_processed.bfill()

# Calculate column means for numeric columns
numeric_cols = df_processed.select_dtypes(include=[np.number]).columns
column_means = {}
for col in numeric_cols:
    # Filter out infinities when calculating means
    filtered_col = df_processed[col].replace([np.inf, -np.inf], np.nan)
    column_means[col] = filtered_col.mean()

# Fill remaining NaNs with calculated means
df_processed = df_processed.fillna(column_means)

# Check for any remaining NaNs
nan_count_after = df_processed.isna().sum().sum()
if nan_count_after > 0:
    print(f"Still have {nan_count_after} NaNs after handling. Replacing with zeros.")
    # Replace any remaining NaNs with zeros as a last resort
    df_processed = df_processed.fillna(0)

# Check for infinity values
inf_count = np.isinf(df_processed.select_dtypes(include=[np.number])).sum().sum()
if inf_count > 0:
    print(f"Found {inf_count} infinity values. Replacing with large finite values.")
    # Replace infinity values with large finite values
    df_processed = df_processed.replace([np.inf, -np.inf], [1e9, -1e9])

# Final NaN check
final_nan_count = df_processed.isna().sum().sum()
print(f"Final NaN values after handling: {final_nan_count}")
```

```python
import torch
import numpy as np
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import StandardScaler
from torch.utils.data import DataLoader, TensorDataset

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

from src.evaluate_metrics import evaluate_with_multiple_metrics
from src.lstm_model import MultiOutputLSTMModel
```

```python
# Check if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
# Combine features for temperature and humidity
combined_features = list(set(temp_feature_names).union(set(humidity_feature_names)))
X_combined = df_copy[combined_features]

# Combine targets for temperature and humidity
y_combined = df_copy[["Indoor_temperature_room", "Humidity"]]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_combined, y_combined, test_size=0.2, random_state=42
)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define batch size
BATCH_SIZE = 64
# Define sequence length for LSTM input
SEQ_LENGTH = 24

# Convert data to PyTorch tensors
X_train_tensor = torch.FloatTensor(X_train_scaled).to(device)
y_train_tensor = torch.FloatTensor(y_train.values).to(device)
X_test_tensor = torch.FloatTensor(X_test_scaled).to(device)
y_test_tensor = torch.FloatTensor(y_test.values).to(device)

# Prepare tensors with shape [batch_size, sequence_length=1, features]
X_train_tensor = X_train_tensor.unsqueeze(1)
X_test_tensor = X_test_tensor.unsqueeze(1)

# Create datasets
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

# Create DataLoaders
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)
```

```python
# Initialize the model
input_size = X_train_scaled.shape[1]
model = MultiOutputLSTMModel(input_size=input_size).to(device)

# Set up loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)

# Learning rate scheduler
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode="min", factor=0.5, patience=5
)
```

```python
# Training function
def train_model(
    model, train_loader, test_loader, optimizer, scheduler, num_epochs=300, patience=15
):
    train_losses = []
    val_losses = []
    best_val_loss = float("inf")
    counter = 0  # Counter for early stopping
    best_model_state = None

    # Initialize mixed precision training (if available)
    use_amp = torch.cuda.is_available()  # Only use AMP with CUDA
    # Mixed Precision Training setup
    scaler = torch.amp.GradScaler()

    for epoch in range(num_epochs):
        # Training phase
        model.train()
        running_loss = 0.0

        for inputs, targets in train_loader:
            optimizer.zero_grad()

            # Use mixed precision for forward and backward pass
            if use_amp:
                with torch.amp.autocast(device_type="cuda"):
                    temp_out, humidity_out = model(inputs)
                    loss_temp = criterion(temp_out, targets[:, 0:1])  # Temperature loss
                    loss_humidity = criterion(
                        humidity_out, targets[:, 1:2]
                    )  # Humidity loss
                    loss = loss_temp + loss_humidity  # Combined loss

                # Scale gradients and call backward
                scaler.scale(loss).backward()

                # Unscale before gradient clipping
                scaler.unscale_(optimizer)

                # Apply gradient clipping to prevent exploding gradients
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

                # Update weights with the scaled gradients
                scaler.step(optimizer)

                # Update the scale for next iteration
                scaler.update()
            else:
                # Regular training (no mixed precision)
                temp_out, humidity_out = model(inputs)
                loss_temp = criterion(temp_out, targets[:, 0:1])
                loss_humidity = criterion(humidity_out, targets[:, 1:2])
                loss = loss_temp + loss_humidity
                loss.backward()

                # Apply gradient clipping
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

                # Regular optimizer step
                optimizer.step()

            running_loss += loss.item() * inputs.size(0)

        epoch_train_loss = running_loss / len(train_loader.dataset)
        train_losses.append(epoch_train_loss)

        # Validation phase
        model.eval()
        running_val_loss = 0.0

        with torch.no_grad():
            for inputs, targets in test_loader:
                # No need for mixed precision in evaluation
                temp_out, humidity_out = model(inputs)
                loss_temp = criterion(temp_out, targets[:, 0:1])
                loss_humidity = criterion(humidity_out, targets[:, 1:2])
                loss = loss_temp + loss_humidity
                running_val_loss += loss.item() * inputs.size(0)

        epoch_val_loss = running_val_loss / len(test_loader.dataset)
        val_losses.append(epoch_val_loss)

        # Learning rate scheduling
        scheduler.step(epoch_val_loss)

        # Print progress
        if (epoch + 1) % 10 == 0:
            print(
                f"Epoch {epoch+1}/{num_epochs}, Train Loss: {epoch_train_loss:.6f}, Val Loss: {epoch_val_loss:.6f}"
            )

        # Early stopping check
        if epoch_val_loss < best_val_loss:
            best_val_loss = epoch_val_loss
            counter = 0
            best_model_state = model.state_dict().copy()
        else:
            counter += 1

        if counter >= patience:
            print(f"Early stopping at epoch {epoch+1}")
            break

    # Load best model
    if best_model_state is not None:
        model.load_state_dict(best_model_state)

    return model, train_losses, val_losses
```

```python
# Train the multi-output model
print("Training Multi-Output Model...")
model, train_losses, val_losses = train_model(
    model, train_loader, test_loader, optimizer, scheduler
)
```

## **8. Environment Simulation Test**
**Steps:**
1. Import environment controller and related libraries
2. Create simulation function to work with different model types
3. Load outdoor data for simulation
4. Run simulations with LSTM model
5. Run simulations with Random Forest model
6. Run simulations with XGBoost model
7. Load and process simulation results
8. Generate comparative visualizations of model predictions
9. Analyze simulation performance differences

```python
from datetime import timedelta, datetime
from src.environment_controller import EnvironmentController


def run_environment_simulation(model_type="lstm"):
    """
    Run the environment simulation with a specified model type.

    Parameters:
    - model_type (str): Type of model to use ('lstm', 'rf', or 'xgb')

    Returns:
    - DataFrame with simulation results
    """
    # Define input features (use the same ones used for training)
    input_features = combined_features

    # Load outdoor data from the new dataset
    outdoor_data = pd.read_csv("Data/processed_data.csv")

    # Make sure DateTime column is there, if not create it
    if "DateTime" not in outdoor_data.columns:
        if "Date" in outdoor_data.columns and "Time" in outdoor_data.columns:
            outdoor_data["DateTime"] = pd.to_datetime(
                outdoor_data["Date"] + " " + outdoor_data["Time"],
                format="%d/%m/%Y %H:%M",
            )
        else:
            # Create a date time column from the first entry starting at midnight
            start_time = datetime.now().replace(
                hour=0, minute=0, second=0, microsecond=0
            )
            date_times = [
                start_time + timedelta(minutes=15 * i) for i in range(len(outdoor_data))
            ]
            outdoor_data["DateTime"] = date_times

    # Make sure all required features are in the dataset
    for feature in input_features:
        if feature not in outdoor_data.columns:
            # If a feature is missing, fill it with a default value
            print(f"Feature {feature} not found in data, using random values")
            outdoor_data[feature] = np.random.rand(len(outdoor_data))

    # Start time (use the first timestamp in the data or default to now)
    # Ensure start_time is a datetime object, not a string
    start_time = pd.to_datetime(
        outdoor_data["DateTime"].min()
        if "DateTime" in outdoor_data.columns
        else datetime.now()
    )

    if model_type.lower() == "lstm":
        # Use the LSTM model
        print("Running simulation with LSTM model...")

        # Initialize the model
        input_size = len(input_features)
        model = MultiOutputLSTMModel(input_size=input_size).to(device)

        # Load the trained model parameters
        model.load_state_dict(
            torch.load("Models/multi_output_model.pth", map_location=device)
        )
        model.eval()

        # Create controller
        controller = EnvironmentController(
            model_path="Models/multi_output_model.pth",
            input_features=input_features,
            model_type="lstm",
        )

    elif model_type.lower() == "rf":
        # Use Random Forest model
        print("Running simulation with Random Forest model...")

        # Load the trained Random Forest models
        rf_temp = joblib.load("Models/rf_temp.joblib")
        rf_humidity = joblib.load("Models/rf_humidity.joblib")

        # Create controller
        controller = EnvironmentController(
            model_type="rf",
            input_features=input_features,
            temp_model=rf_temp,
            humidity_model=rf_humidity,
        )

    elif model_type.lower() == "xgb":
        # Use XGBoost model
        print("Running simulation with XGBoost model...")

        # Load the trained XGBoost models
        xgb_temp = joblib.load("Models/xgb_temp.joblib")
        xgb_humidity = joblib.load("Models/xgb_humidity.joblib")

        # Create controller
        controller = EnvironmentController(
            model_type="xgb",
            input_features=input_features,
            temp_model=xgb_temp,
            humidity_model=xgb_humidity,
        )

    else:
        raise ValueError("Invalid model_type. Choose 'lstm', 'rf', or 'xgb'")

    # Run simulation for 96 intervals (24 hours with 15-minute intervals)
    results = controller.control_environment(
        outdoor_data,
        start_time,
        num_intervals=96,
        output_file=f"Ouput_Data/indoor_environment_control_results_{model_type}.csv",
    )

    print(
        f"Simulation complete with {model_type.upper()} model! First few rows of results:"
    )
    print(results.head())
    return results
```
